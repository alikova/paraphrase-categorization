```python
import torch
from transformers import pipeline

# Function to load model
def load_model():
    model_id = "dvres/GaMS-1B-Translator_0.1"
    translator = pipeline(
        "text-generation",
        model=model_id,
        device_map="auto",
        token="hf"
    )
    return translator

# Batch translation function
def translate_batch(texts, translator):
    translations = []
    for text in texts:
        # Format input for the model
        messages = [{"role": "english", "content": text}]

        # Generate translation
        response = translator(
            messages,
            max_new_tokens=512
        )[0]["generated_text"][-1]["content"]

        translations.append(response)
    return translations

# File paths
input_file_path = "/content/drive/My Drive/Colab Notebooks/msr_paraphrase_data.txt"
output_file_path = "/content/drive/My Drive/Colab Notebooks/translated_msr_paraphrase_data.txt"

# Setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load model
translator = load_model()
print("Model loaded successfully")

# Translation in batches
batch_size = 32  # Adjust based on your GPU memory
batch = []

try:
    with open(input_file_path, "r", encoding="utf-8") as infile, \
         open(output_file_path, "w", encoding="utf-8") as outfile:

        for line_num, line in enumerate(infile, 1):
            line = line.strip()
            if line:  # Ignore empty lines
                batch.append(line)

            # Process full batches
            if len(batch) == batch_size:
                try:
                    translated_batch = translate_batch(batch, translator)
                    outfile.write("\n".join(translated_batch) + "\n")
                    print(f"Processed batch: lines {line_num - batch_size + 1} to {line_num}")
                    batch = []  # Reset batch

                    # Optional: Clear CUDA cache if memory usage is high
                    if torch.cuda.is_available():
                        torch.cuda.empty_cache()

                except Exception as e:
                    print(f"Error processing batch at line {line_num}: {str(e)}")
                    continue

        # Process remaining lines
        if batch:
            try:
                translated_batch = translate_batch(batch, translator)
                outfile.write("\n".join(translated_batch) + "\n")
                print(f"Processed remaining {len(batch)} lines")
            except Exception as e:
                print(f"Error processing final batch: {str(e)}")

    print(f"Translation complete! Translated file saved to: {output_file_path}")

except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
def add_indices_to_pairs(aligned_pairs_path):
    # Read the entire file
    with open(aligned_pairs_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Split the content into pairs based on the separator
    pairs = content.split('---\n')

    # Process each pair and add indices
    indexed_content = ''
    for idx, pair in enumerate(pairs, 1):
        if not pair.strip():  # Skip empty pairs
            continue

        # Add index to the Original line
        if 'Original:' in pair:
            pair = pair.replace('Original:', f'Original [{idx}]:')

        # Add the separator back except for the last pair
        indexed_content += pair + ('---\n' if idx < len(pairs) else '')

    # Write back to file
    with open(aligned_pairs_path + '.indexed', 'w', encoding='utf-8') as file:
        file.write(indexed_content)

    print(f"Added indices to {len(pairs)} pairs")
    return aligned_pairs_path + '.indexed'
```

```python
import torch
from transformers import pipeline
from datasets import Dataset
from tqdm.auto import tqdm
import logging
import time
from concurrent.futures import ThreadPoolExecutor, as_completed
import math
from typing import List, Iterator


# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

def load_model():
    model_id = "dvres/GaMS-1B-Translator_0.1"
    translator = pipeline(
        "text-generation",
        model=model_id,
        device_map="auto",
        token="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"
    )
    return translator

def chunk_file(file_path: str, chunk_size: int = 10000) -> Iterator[List[str]]:
    """Read file in chunks to avoid loading everything into memory"""
    current_chunk = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            current_chunk.append(line.strip())
            if len(current_chunk) == chunk_size:
                yield current_chunk
                current_chunk = []
        if current_chunk:
            yield current_chunk

def translate_batch(batch: List[str], translator, batch_size: int = 32) -> List[str]:
    """Translate a batch of texts"""
    translations = []

    # Process mini-batches within the chunk
    for i in range(0, len(batch), batch_size):
        mini_batch = batch[i:i + batch_size]
        try:
            batch_translations = []
            for text in mini_batch:
                messages = [{"role": "english", "content": text}]
                response = translator(
                    messages,
                    max_new_tokens=512
                )[0]["generated_text"][-1]["content"]
                batch_translations.append(response)
            translations.extend(batch_translations)

            # Clear CUDA cache periodically
            if torch.cuda.is_available() and i % (batch_size * 5) == 0:
                torch.cuda.empty_cache()

        except Exception as e:
            logger.error(f"Error in mini-batch translation: {str(e)}")
            translations.extend(["ERROR: Translation failed"] * len(mini_batch))

    return translations

def translate_chunk(chunk: List[str], chunk_id: int, translator, output_path: str, batch_size: int):
    """Translate a chunk of text and save to a temporary file"""
    temp_output_path = f"{output_path}.part{chunk_id}"
    translations = translate_batch(chunk, translator, batch_size)

    with open(temp_output_path, 'w', encoding='utf-8') as f:
        f.write('\n'.join(translations) + '\n')

    return temp_output_path, len(translations)

def merge_output_files(temp_files: List[str], final_output_path: str):
    """Merge all temporary files into the final output file"""
    with open(final_output_path, 'w', encoding='utf-8') as outfile:
        for temp_file in sorted(temp_files):
            with open(temp_file, 'r', encoding='utf-8') as infile:
                outfile.write(infile.read())
            os.remove(temp_file)  # Clean up temp file

def main():
    # File paths
    input_path = "/content/drive/My Drive/Colab Notebooks/para-nmt-50m-small.txt"
    output_path = "/content/drive/My Drive/Colab Notebooks/translated_para-nmt-50m-small.txt"

    # Parameters
    chunk_size = 10000  # Process 10,000 lines at a time
    batch_size = 32     # GPU batch size
    num_workers = 4     # Number of parallel workers

    # Setup
    start_time = time.time()
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    logger.info(f"Using device: {device}")

    # Load model
    translator = load_model()
    logger.info("Model loaded successfully")

    # Process file in chunks
    temp_files = []
    total_lines = sum(1 for _ in open(input_path, 'r', encoding='utf-8'))

    with ThreadPoolExecutor(max_workers=num_workers) as executor:
        futures = []

        # Submit translation tasks for each chunk
        for chunk_id, chunk in enumerate(chunk_file(input_path, chunk_size)):
            future = executor.submit(
                translate_chunk,
                chunk,
                chunk_id,
                translator,
                output_path,
                batch_size
            )
            futures.append(future)

        # Monitor progress
        with tqdm(total=total_lines, desc="Translating") as pbar:
            for future in as_completed(futures):
                temp_file, lines_processed = future.result()
                temp_files.append(temp_file)
                pbar.update(lines_processed)

    # Merge results
    logger.info("Merging output files...")
    merge_output_files(temp_files, output_path)

    end_time = time.time()
    logger.info(f"Translation completed in {end_time - start_time:.2f} seconds")

if __name__ == "__main__":
    main()
```

```python
# new code for more memory-efficient approach using iterators to save the matching pairs

import torch
import numpy as np
from transformers import pipeline

def load_model():
    model_id = "dvres/GaMS-1B-Translator_0.1"
    translator = pipeline(
        "text-generation",
        model=model_id,
        device_map="auto",
        token="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"
    )
    return translator

def translate_batch(texts, translator):
    translations = []
    for text in texts:
        messages = [{"role": "english", "content": text}]
        response = translator(
            messages,
            max_new_tokens=512
        )[0]["generated_text"][-1]["content"]
        translations.append(response)
    return translations

def select_and_translate(input_file_path, translator, dataset_name, n_samples=10000, batch_size=32):
    with open(input_file_path, "r", encoding="utf-8") as f:
        total_lines = sum(1 for _ in f)

    selected_indices = set(np.random.choice(total_lines, n_samples, replace=False))
    selected_texts = []

    with open(input_file_path, "r", encoding="utf-8") as f:
        for i, line in enumerate(f):
            if i in selected_indices:
                selected_texts.append(line.strip())
                if len(selected_texts) == batch_size:
                    try:
                        translations = translate_batch(selected_texts, translator)
                        save_pairs(selected_texts, translations, dataset_name)
                        print(f"Processed batch: {len(selected_texts)} lines")
                        selected_texts = []
                        if torch.cuda.is_available():
                            torch.cuda.empty_cache()
                    except Exception as e:
                        print(f"Error processing batch: {str(e)}")
                        continue

    if selected_texts:
        try:
            translations = translate_batch(selected_texts, translator)
            save_pairs(selected_texts, translations)
            print(f"Processed remaining {len(selected_texts)} lines")
        except Exception as e:
            print(f"Error processing final batch: {str(e)}")

def save_pairs(originals, translations, dataset_name):
    with open(f"/content/drive/My Drive/Colab Notebooks/{dataset_name}_selected_originals.txt", "a", encoding="utf-8") as f1, \
         open(f"/content/drive/My Drive/Colab Notebooks/{dataset_name}_selected_translations.txt", "a", encoding="utf-8") as f2, \
         open(f"/content/drive/My Drive/Colab Notebooks/{dataset_name}_aligned_pairs.txt", "a", encoding="utf-8") as f3:
        for orig, trans in zip(originals, translations):
            f1.write(orig + "\n")
            f2.write(trans + "\n")
            f3.write(f"Original: {orig}\nTranslation: {trans}\n---\n")
```
